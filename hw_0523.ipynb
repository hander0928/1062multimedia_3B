```python
from skimage import io
import matplotlib.pyplot as plt
import numpy as np
```

```python
def reduceThree(image):
    rows,cols,dims=image.shape
    ret_img = np.uint8(np.zeros((rows,cols,3)))
    for i in range(0,rows):
        for j in range(0,cols):
            luminance = 0
            for k in range(0,3):
                luminance = luminance + image[i,j,k]
            luminance = luminance/3
            if luminance < 80:
                ret_img[i,j,0] = 0
                ret_img[i,j,1] = 0
                ret_img[i,j,2] = 0
            elif luminance > 150:
                ret_img[i,j,0] = 255
                ret_img[i,j,1] = 255
                ret_img[i,j,2] = 255
            else:
                ret_img[i,j,0] = 255
                ret_img[i,j,1] = 255
                ret_img[i,j,2] = 0
    return ret_img

img=io.imread('fish.jpg')
io.imshow(img)

img2 = reduceThree(img)
plt.figure()
io.imshow(img2)
io.show()
```
